```python
## add your code here
#include <vector>
#include <set>
#include <map>
#include <queue>
#include <string>
#include <algorithm>
#include <iostream>
#include <bitset>
#include <functional>
#include <numeric>
#include <cstdio>
#include <cstring>
#include <cstdlib>
#include <cassert>
#include <cmath>
#include <ctime>
#include <complex>
using namespace std;
 
typedef long long ll;
typedef complex<double> comp;
 
#define fr first
#define sc second
#define ad push_back
#define em emplace_back
 
const int BUF_MAX = 131072;
static char buffer[BUF_MAX];
static char *hd = buffer, *tl = buffer;
 
inline char gc() {
    if (hd == tl) {
        tl = buffer + fread(buffer, 1, BUF_MAX, stdin);
        hd = buffer;
        if (hd == tl) return EOF;
    }
    return *hd++;
}
 
inline int read() {
    int cur = 0;
    char c = gc();
    while (c < 48) c = gc();
    while (c > 47) {
        cur = cur * 10 + (c & 15);
        c = gc();
    }
    return cur;
}
 
inline int fpow(int bs, int ex, int md) {
    int r = 1;
    while (ex) {
        if (ex & 1) r = (ll)r * bs % md;
        bs = (ll)bs * bs % md;
        ex >>= 1;
    }
    return r;
}
 
const int MAX_ARR = 1024;
const int LOG_UP = 20;
const int MOD = 1000000007;
const int INF = 1000000007;
int dx[] = {1, 0, -1, 0};
int dy[] = {0, 1, 0, -1};
 
int val[MAX_ARR];
int loc[MAX_ARR];
vector<int> ops;
int n, x, y, k;
 
void upd() {
    for (int i = 0; i < n; ++i) loc[val[i]] = i;
}
 
void m_swap() {
    ops.ad(0);
    for (int i = 0; i < n; ++i) {
        if (val[i] == x) val[i] = y;
        else if (val[i] == y) val[i] = x;
    }
    upd();
}
 
void add(int v) {
    if (!v) return;
    ops.ad(v);
    for (int i = 0; i < n; ++i) val[i] = (val[i] + v) % n;
    upd();
}
 
void xr(int v) {
    if (!v) return;
    ops.ad(-v);
    for (int i = 0; i < n; ++i) val[i] ^= v;
    upd();
}
 
void param(int a, int b, int &p1, int &p2) {
    int d = (b - a + n - k + n) % n;
    p1 = p2 = 0;
    for (int s = n >> 1; s >= (k << 1); s >>= 1) {
        if (d >= s) {
            d -= s;
            p2 += s >> 1;
        } else p1 += s >> 1;
    }
    p1 += n >> 1;
    p1 += a & (k - 1);
    p2 += a & (k - 1);
}
 
void swp(int c, int d) {
    if ((c / k & 1) == (d / k & 1)) {
        int mid;
        if (!(c / k & 1)) mid = (c & (k - 1)) + k;
        else mid = c & (k - 1);
        swp(c, mid);
        swp(d, mid);
        swp(c, mid);
    } else {
        int pa, pb;
        param(x, y, pa, pb);
        int pc, pd;
        param(c, d, pc, pd);
        add((pc - c + n) % n);
        xr(pc ^ pa);
        add((x - pa + n) % n);
        m_swap();
        add((pa - x + n) % n);
        xr(pc ^ pa);
        add((c - pc + n) % n);
    }
}
 
struct P {
    int v[MAX_ARR];
    int s;
    vector<int> op;
 
    bool operator < (const P& t) const {
        for (int i = 0; i < s; ++i)
            if (v[i] != t.v[i]) return v[i] < t.v[i];
        return false;
    }
 
    void pr() const {
        for (int i = 0; i < s; ++i)
            printf("%d%c", v[i], " \n"[i == s - 1]);
    }
 
    P inv() const {
        P r;
        for (int i = 0; i < s; ++i) r.v[v[i]] = i;
        return r;
    }
 
    bool ok() const {
        for (int i = 1; i < s; ++i)
            if (v[i] < v[i - 1]) return false;
        return true;
    }
 
    bool run() {
        bool vis[100005] = {0};
        for (int i = 0; i < s; ++i) vis[i] = 1;
        for (int i = 0; i < s; ++i) if (!vis[i]) return false;
        if (s == 1) return true;
 
        P l, r;
        l.s = r.s = s / 2;
        for (int i = 0; i < s / 2; ++i) {
            l.v[i] = v[i * 2] / 2;
            r.v[i] = v[i * 2 + 1] / 2;
        }
 
        if (!l.run() || !r.run()) return false;
        if (v[0] & 1) op.ad(s == 2 ? 1 : -1);
 
        int ta = 0;
        for (int o : l.op) {
            if (o > 0) op.ad(-1), op.ad(1);
            else op.ad(o * 2), ta ^= (-o) * 2;
        }
        if (ta) op.ad(-ta);
 
        int tb = 0;
        for (int o : r.op) {
            if (o > 0) op.ad(1), op.ad(-1);
            else op.ad(o * 2), tb ^= (-o) * 2;
        }
 
        if ((tb & (s / 2)) != (ta & (s / 2))) {
            assert(0);
            for (int i = 0; i < s / 4; ++i) op.ad(-1), op.ad(1);
        }
 
        if (ta >= s / 2) ta -= s / 2;
        if (tb >= s / 2) tb -= s / 2;
        if (ta != tb) return false;
 
        vector<int> tmp;
        for (int o : op) {
            if (tmp.empty()) tmp.ad(o);
            else {
                if (o < 0 && tmp.back() < 0) {
                    tmp.back() = -((-tmp.back()) ^ (-o));
                    if (!tmp.back()) tmp.pop_back();
                } else tmp.ad(o);
            }
        }
        swap(tmp, op);
        return true;
    }
};
 
int bak[MAX_ARR];
 
int main() {
    n = read();
    x = read();
    y = read();
 
    for (int i = 0; i < n; ++i) val[i] = read();
    upd();
 
    k = (x - y + n) % n;
    k &= -k;
    if (!k) k = n;
 
    if (k > 1) {
        P p;
        p.s = k;
        for (int i = 0; i < n; ++i) p.v[i] = val[i] & (k - 1);
        if (!p.run()) {
            puts("-1");
            return 0;
        }
        for (int o : p.op) {
            if (o > 0) add(o);
            else xr(-o);
        }
    }
 
    for (int i = 0; i < k; ++i) {
        vector<int> vec;
        for (int j = i; j < n; j += k) vec.ad(val[j]);
        sort(vec.begin(), vec.end());
        bool f = 1;
        int ptr = 0;
        for (int j = i; j < n; j += k, ptr++) {
            if (vec[ptr] != j) {
                f = 0;
                break;
            }
        }
        if (!f) {
            puts("-1");
            return 0;
        }
        for (int j = i; j < n; j += k) {
            if (val[j] != j) swp(j, val[j]);
        }
    }
 
    for (int i = 0; i < n; ++i) assert(val[i] == i);
    printf("%d\n", (int)ops.size());
    for (int o : ops) {
        if (!o) puts("0");
        else if (o < 0) printf("1 %d\n", -o);
        else printf("2 %d\n", o);
    }
 
    return 0;
}
```

## B 长跑

```python
## add your code here
'''
解题思路：
本题的目标是判断小明是否能在金币限制 S 和体力上限 Maxn 的约束下到达终点 L。
本质是一个带有权重的路径可达性问题，由于 N 较小（2000），可以使用动态规划（DP）解决。

Step 1. 环境准备与预处理
- 题目包含多组数据，需使用 while(cin >> ...) 持续读入。
- 初始判断：如果 Maxn >= L，小明无需补给即可到达，直接输出 "Yes"。
- 排序：由于补给站 Pi 给出的顺序可能乱序，必须先按位置 Pi 从小到大排序，以便进行线性 DP 转移。

Step 2. 状态定义
- 定义 dp[i] 为：到达第 i 个补给站并在该站“完成体力补满”所需要花费的最少金币数。
- 初始化 dp 数组为无穷大（INF），表示该站初始不可达。

Step 3. 状态转移方程 (O(N^2))
对于每一个补给站 i (0 到 N-1)：
1. 起点直达：检查是否能从起点 (0) 直接跑到站 i。
   如果 Pi <= Maxn，则 dp[i] = Ci。
2. 站点中转：遍历所有在 i 之前的站点 j (0 到 i-1)。
   如果小明能在第 j 站补满体力后跑到第 i 站（即 Pi - Pj <= Maxn），
   且第 j 站是可达的（dp[j] <= S），
   则更新 dp[i] = min(dp[i], dp[j] + Ci)。

Step 4. 终点判定
- 在计算每个 dp[i] 时，实时检查：
  如果 dp[i] <= S（费用在预算内）且从该站补满体力后能直接冲到终点（L - Pi <= Maxn），
  则标记为成功，输出 "Yes"。
- 若遍历完所有站点仍无法到达终点，输出 "No"。

注意点：
- 费用 Ci >= 1000 且 S <= 2000 意味着小明全程最多补给 2 次。
- 必须严格遵守输出大小写（Yes/No），否则 OJ 会判定错误。
'''
#include <iostream>
#include <vector>
#include <algorithm>

using namespace std;

// 使用 long long 防止费用累加时可能的溢出
typedef long long ll;

struct Station {
    int p; // 补给站位置
    int c; // 补给站花费
};

// 排序规则：按位置从小到大排序
bool compareStations(const Station& a, const Station& b) {
    if (a.p != b.p) return a.p < b.p;
    return a.c < b.c; // 位置相同时，便宜的优先
}

void solve() {
    int N, L, Maxn, S;
    // 处理多组数据读入
    while (cin >> N >> L >> Maxn >> S) {
        vector<Station> stations(N);
        for (int i = 0; i < N; ++i) {
            cin >> stations[i].p >> stations[i].c;
        }

        // 1. 特殊情况：如果初始体力直接能到终点
        if (Maxn >= L) {
            cout << "Yes" << endl;
            continue;
        }

        // 2. 如果没有补给站且跑不到终点
        if (N == 0) {
            cout << "No" << endl;
            continue;
        }

        // 3. 对补给站进行排序
        sort(stations.begin(), stations.end(), compareStations);

        // 4. 动态规划
        const ll INF = 1e16; 
        vector<ll> dp(N, INF);
        bool success = false;

        for (int i = 0; i < N; ++i) {
            // A. 尝试直接从起点（位置0）出发到站 i
            if (stations[i].p <= Maxn) {
                dp[i] = (ll)stations[i].c;
            }

            // B. 尝试从之前的站点 j 中转到站 i
            for (int j = 0; j < i; ++j) {
                // 如果 j 站可达，且 j 到 i 的距离在体力范围内
                if (dp[j] <= (ll)S && (stations[i].p - stations[j].p) <= Maxn) {
                    if (dp[j] + (ll)stations[i].c < dp[i]) {
                        dp[i] = dp[j] + (ll)stations[i].c;
                    }
                }
            }

            // 5. 判定：当前站补满后能否到达终点
            if (dp[i] <= (ll)S && (L - stations[i].p) <= Maxn) {
                success = true;
                break; 
            }
        }

        if (success) {
            cout << "Yes" << endl;
        } else {
            cout << "No" << endl;
        }
    }
}

int main() {
    // 提升 IO 效率，应对大规模数据
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);

    solve();

    return 0;
}
```

## C 最长回文

```python
#include <iostream>
#include <vector>
#include <string>
#include <algorithm>

using namespace std;

// 使用 64 位无符号整数自然溢出作为哈希模数
typedef unsigned long long ull;
const ull HASH_BASE = 911382323ULL;

// 结构体用于存储候选回文片段
struct Candidate {
    int max_len_possible;
    int pos_A_rev;
    int b_start;
    int base_len;
    int limit;

    // 用于降序排序：max_len_possible 大的排在前面
    bool operator<(const Candidate& other) const {
        return max_len_possible > other.max_len_possible;
    }
};

// Manacher 算法同时计算奇回文和偶回文半径
void manacher_radius(const string& s, vector<int>& odd_radius, vector<int>& even_radius) {
    int length = s.length();
    odd_radius.assign(length, 0);
    even_radius.assign(length, 0);

    // 1. 奇回文半径
    int l = 0, r = -1;
    for (int center = 0; center < length; ++center) {
        int k = (center > r) ? 1 : min(odd_radius[l + r - center], r - center + 1);
        while (center - k >= 0 && center + k < length && s[center - k] == s[center + k]) {
            k++;
        }
        odd_radius[center] = k;
        if (center + k - 1 > r) {
            l = center - k + 1;
            r = center + k - 1;
        }
    }

    // 2. 偶回文半径
    l = 0, r = -1;
    for (int center = 0; center < length; ++center) {
        int k = (center > r) ? 0 : min(even_radius[l + r - center + 1], r - center + 1);
        while (center - k - 1 >= 0 && center + k < length && s[center - k - 1] == s[center + k]) {
            k++;
        }
        even_radius[center] = k;
        if (center + k - 1 > r) {
            l = center - k;
            r = center + k - 1;
        }
    }
}

int main() {
    // 优化输入输出性能
    ios_base::sync_with_stdio(false);
    cin.tie(NULL);

    int n;
    if (!(cin >> n)) return 0;

    string str_A, str_B;
    cin >> str_A >> str_B;

    vector<int> odd_radius_A, even_radius_A;
    vector<int> odd_radius_B, even_radius_B;

    manacher_radius(str_A, odd_radius_A, even_radius_A);
    manacher_radius(str_B, odd_radius_B, even_radius_B);

    int max_palindrome_len = 1;

    // 先计算单字符串内部的最大回文长度
    for (int rad : odd_radius_A)  max_palindrome_len = max(max_palindrome_len, 2 * rad - 1);
    for (int rad : odd_radius_B)  max_palindrome_len = max(max_palindrome_len, 2 * rad - 1);
    for (int rad : even_radius_A) max_palindrome_len = max(max_palindrome_len, 2 * rad);
    for (int rad : even_radius_B) max_palindrome_len = max(max_palindrome_len, 2 * rad);

    // 字符串 A 的反转
    string str_A_rev = str_A;
    reverse(str_A_rev.begin(), str_A_rev.end());

    // 预处理字符串哈希和 Base 幂次
    vector<ull> power_base(n + 1, 1);
    vector<ull> hash_A_rev(n + 1, 0);
    vector<ull> hash_B(n + 1, 0);

    for (int i = 0; i < n; ++i) {
        power_base[i + 1] = power_base[i] * HASH_BASE;
    }

    ull h_val = 0;
    for (int i = 0; i < n; ++i) {
        h_val = h_val * HASH_BASE + str_A_rev[i];
        hash_A_rev[i + 1] = h_val;
    }

    h_val = 0;
    for (int i = 0; i < n; ++i) {
        h_val = h_val * HASH_BASE + str_B[i];
        hash_B[i + 1] = h_val;
    }

    vector<Candidate> candidates;

    // lambda 表达式：添加候选点
    auto add_candidate = [&](int a_end, int b_start, int base_len) {
        if (a_end < 0 || b_start >= n) return;

        int limit = min(a_end + 1, n - b_start);
        if (limit <= 0) return;

        int max_len_possible = base_len + (limit << 1);
        if (max_len_possible <= max_palindrome_len) return;

        int pos_in_A_rev = n - 1 - a_end;
        candidates.push_back({max_len_possible, pos_in_A_rev, b_start, base_len, limit});
    };

    // 1. 奇偶回文中心在 A 中
    for (int center = 0; center < n; ++center) {
        add_candidate(center - odd_radius_A[center], center + odd_radius_A[center] - 1, (odd_radius_A[center] << 1) - 1);
        if (even_radius_A[center] > 0) {
            add_candidate(center - even_radius_A[center] - 1, center + even_radius_A[center] - 1, even_radius_A[center] << 1);
        }
    }

    // 2. 奇偶回文中心在 B 中
    for (int center = 0; center < n; ++center) {
        add_candidate(center - odd_radius_B[center] + 1, center + odd_radius_B[center], (odd_radius_B[center] << 1) - 1);
        if (even_radius_B[center] > 0) {
            add_candidate(center - even_radius_B[center], center + even_radius_B[center], even_radius_B[center] << 1);
        }
    }

    // 3. 纯拼接点（不以现有回文为中心，直接从边界扩展）
    for (int k = 0; k < n; ++k) {
        add_candidate(k, k, 0);
    }

    // 优先处理理论可能得到更长答案的候选
    sort(candidates.begin(), candidates.end());

    // lambda 表达式：利用 O(1) 关键哈希比对子串是否相同
    auto hashes_equal = [&](int pos_A_rev, int pos_B, int length) {
        ull hash1 = hash_A_rev[pos_A_rev + length] - hash_A_rev[pos_A_rev] * power_base[length];
        ull hash2 = hash_B[pos_B + length] - hash_B[pos_B] * power_base[length];
        return hash1 == hash2;
    };

    // 遍历候选集
    for (const auto& cand : candidates) {
        if (cand.max_len_possible <= max_palindrome_len) {
            break; // 剪枝：后续的理论最大值都不可能超越当前最优解
        }

        int need_extend = ((max_palindrome_len - cand.base_len) >> 1) + 1;
        if (need_extend > cand.limit) continue;
        if (str_A_rev[cand.pos_A_rev] != str_B[cand.b_start]) continue;
        if (!hashes_equal(cand.pos_A_rev, cand.b_start, need_extend)) continue;

        // 倍增思想配合二分细化，加速寻找最大扩展长度
        int lo = need_extend;
        if (lo < cand.limit) {
            int nxt = lo + 1;
            if (hashes_equal(cand.pos_A_rev, cand.b_start, nxt)) {
                lo = nxt;
                int step = 2;
                while (lo + step <= cand.limit && hashes_equal(cand.pos_A_rev, cand.b_start, lo + step)) {
                    lo += step;
                    step <<= 1;
                }
                int hi = min(lo + step - 1, cand.limit);
                while (lo < hi) {
                    int mid = (lo + hi + 1) >> 1;
                    if (hashes_equal(cand.pos_A_rev, cand.b_start, mid)) {
                        lo = mid;
                    } else {
                        hi = mid - 1;
                    }
                }
            }
        }

        int candidate_len = cand.base_len + (lo << 1);
        if (candidate_len > max_palindrome_len) {
            max_palindrome_len = candidate_len;
        }
    }

    cout << max_palindrome_len << "\n";

    return 0;
}
```

## D 优惠券

```python
## add your code here
import sys

class BinaryIndexedTree:
    """封装树状数组，用于快速查询和更新可用问号的位置"""
    def __init__(self, size):
        self.size = size
        self.tree_array = [0] * (size + 1)

    def update(self, index, delta):
        while index <= self.size:
            self.tree_array[index] += delta
            index += index & (-index)

    def query_prefix(self, index):
        total = 0
        while index > 0:
            total += self.tree_array[index]
            index -= index & (-index)
        return total

    def find_kth_element(self, k):
        """利用倍增思想在树状数组上二分，寻找第 k 个元素的位置"""
        current_pos = 0
        step = 1 << (self.size.bit_length() - 1)

        while step > 0:
            next_pos = current_pos + step
            if next_pos <= self.size and self.tree_array[next_pos] < k:
                current_pos = next_pos
                k -= self.tree_array[next_pos]
            step >>= 1

        return current_pos + 1


def process_test_case(total_ops, actions):
    MAX_ITEM_ID = 100005
    
    # 状态记录数组
    has_item = [False] * MAX_ITEM_ID
    last_seen_at = [0] * MAX_ITEM_ID
    
    bit = BinaryIndexedTree(total_ops)

    for step in range(1, total_ops + 1):
        action = actions[step - 1]
        op_type = action[0]

        # 遇到问号，存入可用池
        if op_type == b"?":
            bit.update(step, 1)
            continue

        item_id = int(action[1])
        is_buying = (op_type == b"I")

        # 核心逻辑合并：如果当前想“买”但已经“有”了，或者想“卖”但其实“没有”
        # 这就意味着逻辑冲突，需要动用以前的问号来修正历史
        if has_item[item_id] == is_buying:
            left_bound = last_seen_at[item_id]
            
            # 检查在 [left_bound, step] 期间是否有攒下来的问号
            available_wildcards = bit.query_prefix(step) - bit.query_prefix(left_bound - 1)
            
            # 如果没有可用的问号进行时光倒流，当前步骤就是错误发生的地方
            if available_wildcards == 0:
                return step

            # 寻找在 left_bound 之后出现的第一个问号并消耗掉
            needed_kth = bit.query_prefix(left_bound - 1) + 1
            wildcard_pos = bit.find_kth_element(needed_kth)
            bit.update(wildcard_pos, -1)

        # 更新物品状态和最新操作位置
        has_item[item_id] = is_buying
        last_seen_at[item_id] = step

    # 顺利执行完所有操作
    return -1


def main():
    raw_data = sys.stdin.buffer.read().split()
    if not raw_data:
        return
        
    data_iterator = iter(raw_data)
    results = []

    while True:
        try:
            m_str = next(data_iterator)
        except StopIteration:
            # 数据流结束
            break
            
        m = int(m_str)
        actions = []
        
        # 解析每个 Test Case 的操作指令
        for _ in range(m):
            op = next(data_iterator)
            if op == b"?":
                actions.append((op,))
            else:
                target_item = next(data_iterator)
                actions.append((op, target_item))

        # 执行求解并将结果转换为字符串存入数组
        results.append(str(process_test_case(m, actions)))

    # 统一输出
    print("\n".join(results))

if __name__ == "__main__":
    main()
```

## E 任意点

```python
## add your code here
import sys
from collections import deque

def solve_graph_components():
    # 统一读取所有输入项
    input_tokens = sys.stdin.buffer.read().split()
    if not input_tokens:
        return
    
    total_points = int(input_tokens[0])
    token_index = 1

    coord_pairs = []
    distinct_x = set()
    distinct_y = set()

    for _ in range(total_points):
        x_val = int(input_tokens[token_index])
        y_val = int(input_tokens[token_index + 1])
        token_index += 2

        coord_pairs.append((x_val, y_val))
        distinct_x.add(x_val)
        distinct_y.add(y_val)

    # 字典映射进行坐标离散化
    x_mapping = {value: idx for idx, value in enumerate(sorted(distinct_x))}
    y_mapping = {value: idx + len(x_mapping) for idx, value in enumerate(sorted(distinct_y))}

    graph_size = len(x_mapping) + len(y_mapping)
    adjacency_list = [[] for _ in range(graph_size)]

    # 构建二分图的边
    for x_val, y_val in coord_pairs:
        u_node = x_mapping[x_val]
        v_node = y_mapping[y_val]
        adjacency_list[u_node].append(v_node)
        adjacency_list[v_node].append(u_node)

    visited_nodes = [False] * graph_size
    total_components = 0

    # 用 BFS 遍历图的连通块
    for start_node in range(graph_size):
        if not visited_nodes[start_node]:
            total_components += 1
            
            queue = deque([start_node])
            visited_nodes[start_node] = True
            
            while queue:
                current_node = queue.popleft()
                for neighbor in adjacency_list[current_node]:
                    if not visited_nodes[neighbor]:
                        visited_nodes[neighbor] = True
                        queue.append(neighbor)

    # 连通整个图所需的最少连边数
    print(total_components - 1)


if __name__ == "__main__":
    solve_graph_components()
```

## F 通配符匹配

```python
## add your code here
import sys

class StringHasher:
    """将滚动哈希相关的操作封装成类，隐藏底层的位运算细节"""
    def __init__(self, text, base=13331, mask=(1 << 64) - 1):
        self.base = base
        self.mask = mask
        n = len(text)
        self.prefix_hash = [0] * (n + 1)
        for i in range(n):
            self.prefix_hash[i + 1] = (self.prefix_hash[i] * base + text[i]) & mask

    @staticmethod
    def compute_power_base(max_len, base=13331, mask=(1 << 64) - 1):
        powers = [1] * (max_len + 1)
        for i in range(1, max_len + 1):
            powers[i] = (powers[i - 1] * base) & mask
        return powers

    def fetch_sub_hash(self, start, length, powers):
        end = start + length
        return (self.prefix_hash[end] - self.prefix_hash[start] * powers[length]) & self.mask


def extract_segments_blocks(pattern_segment, base=13331, mask=(1 << 64) - 1):
    """解析没有星号的片段，按问号分割出纯字母子块"""
    blocks = []
    length = len(pattern_segment)
    i = 0
    while i < length:
        if pattern_segment[i] == 63:  # 63 是 b'?' 的 ASCII 码
            i += 1
            continue
        
        start_idx = i
        current_hash = 0
        while i < length and pattern_segment[i] != 63:
            current_hash = (current_hash * base + pattern_segment[i]) & mask
            i += 1
        blocks.append((start_idx, i - start_idx, current_hash))
    return blocks


def solve_wildcard_matching():
    tokens = sys.stdin.buffer.read().split()
    if not tokens:
        return
    
    pattern_string = tokens[0]
    total_files = int(tokens[1])
    file_list = tokens[2:2 + total_files]

    # 按星号拆分模式串
    segments = pattern_string.split(b"*")
    star_count = len(segments) - 1
    segment_count = len(segments)

    # 预计算后缀剩余的最短理论长度
    suffix_min_lens = [0] * segment_count
    suffix_min_lens[-1] = len(segments[-1])
    for i in range(segment_count - 2, -1, -1):
        suffix_min_lens[i] = suffix_min_lens[i + 1] + len(segments[i])

    HASH_BASE = 13331
    HASH_MASK = (1 << 64) - 1

    # 对所有片段中的子块预先做哈希计算
    parsed_segments = [extract_segments_blocks(seg, HASH_BASE, HASH_MASK) for seg in segments]
    
    max_len_among_files = max((len(f) for f in file_list), default=0)
    powers_table = StringHasher.compute_power_base(max_len_among_files, HASH_BASE, HASH_MASK)

    output_results = []

    for filename in file_list:
        file_len = len(filename)
        
        if file_len < suffix_min_lens[0]:
            output_results.append("NO")
            continue

        hasher = StringHasher(filename, HASH_BASE, HASH_MASK)

        def verify_match_at(seg_index, text_position):
            """校验特定的片段是否能在文本的指定位置对齐匹配"""
            for offset, length, target_hash in parsed_segments[seg_index]:
                if hasher.fetch_sub_hash(text_position + offset, length, powers_table) != target_hash:
                    return False
            return True

        # 无星号的特殊情况：严格等长匹配
        if star_count == 0:
            if file_len == len(segments[0]) and verify_match_at(0, 0):
                output_results.append("YES")
            else:
                output_results.append("NO")
            continue

        # 有星号的情况：优先验证硬性要求的头部与尾部
        if not verify_match_at(0, 0):
            output_results.append("NO")
            continue

        tail_start_pos = file_len - len(segments[-1])
        if not verify_match_at(star_count, tail_start_pos):
            output_results.append("NO")
            continue

        # 中间夹着的片段采用贪心策略从左往右扫描
        current_search_start = len(segments[0])
        match_success = True

        for seg_idx in range(1, star_count):
            current_seg_len = len(segments[seg_idx])
            latest_start_pos = file_len - suffix_min_lens[seg_idx]
            
            found_position = False
            for pos in range(current_search_start, latest_start_pos + 1):
                if verify_match_at(seg_idx, pos):
                    found_position = True
                    current_search_start = pos + current_seg_len
                    break
            
            if not found_position:
                match_success = False
                break

        output_results.append("YES" if match_success else "NO")

    sys.stdout.write("\n".join(output_results))


if __name__ == "__main__":
    solve_wildcard_matching()
```

## G 汉诺塔

```python
## add your code here
import sys

def solve_hanoi_rules():
    tokens = sys.stdin.buffer.read().split()
    if not tokens:
        return
    
    disk_count = int(tokens[0])
    # 建立移动动作到优先级权重的直接映射
    priority_ranks = {tokens[i]: i - 1 for i in range(1, 7)}
    
    # 柱子映射: 0->A, 1->B, 2->C
    peg_chars = [b"A", b"B", b"C"]
    optimal_target = [0] * 3
    
    # 事先推导 1 号单盘在各柱子时的首选去向
    for src in range(3):
        choice_1 = (src + 1) % 3
        choice_2 = (src + 2) % 3
        
        if priority_ranks[peg_chars[src] + peg_chars[choice_1]] < priority_ranks[peg_chars[src] + peg_chars[choice_2]]:
            optimal_target[src] = choice_1
        else:
            optimal_target[src] = choice_2
            
    # dp_steps[i][j]: i 个盘子在 j 柱组合成整塔所需步数
    # target_location[i][j]: 最终整塔会停留在哪根柱子上
    dp_steps = [[0] * 3 for _ in range(disk_count + 1)]
    target_location = [[0] * 3 for _ in range(disk_count + 1)]
    
    for p in range(3):
        dp_steps[1][p] = 1
        target_location[1][p] = optimal_target[p]
        
    # 自底向上递推多盘子场景
    for k in range(2, disk_count + 1):
        for start_peg in range(3):
            pos_large = start_peg
            pos_small_tower = start_peg
            accumulated_steps = 0
            
            while True:
                accumulated_steps += dp_steps[k - 1][pos_small_tower]
                pos_small_tower = target_location[k - 1][pos_small_tower]
                
                if pos_small_tower == pos_large:
                    target_location[k][start_peg] = pos_small_tower
                    dp_steps[k][start_peg] = accumulated_steps
                    break
                    
                # 最大盘发生转移，移向唯一的空柱
                pos_large = 3 - pos_large - pos_small_tower
                accumulated_steps += 1
                
    print(dp_steps[disk_count][0])

if __name__ == "__main__":
    solve_hanoi_rules()
```

## H 马步距离

```python
## add your code here
import sys

def compute_knight_distance():
    raw_input = sys.stdin.read().split()
    if not raw_input:
        return
    
    x1, y1, x2, y2 = map(int, raw_input)
    dist_x, dist_y = abs(x1 - x2), abs(y1 - y2)
    
    # 动态解包，确保 high 代表较长轴距离，low 代表较短轴距离
    high, low = (dist_x, dist_y) if dist_x > dist_y else (dist_y, dist_x)
    
    # 边缘局限特判
    if high == 1 and low == 0:
        print(3)
        return
    if high == 2 and low == 2:
        print(4)
        return
        
    # 基于棋盘移动特性的理论步数下界计算
    limit_by_axis = (high + 1) >> 1
    limit_by_sum = (high + low + 2) // 3
    
    total_moves = max(limit_by_axis, limit_by_sum)
    
    # 奇偶校验：总步数与总坐标偏移的奇偶性必须契合（用异或位运算改写）
    if (total_moves ^ high ^ low) & 1:
        total_moves += 1
        
    print(total_moves)

if __name__ == "__main__":
    compute_knight_distance()
```

## I 直方图最大矩形

```python
## add your code here
#
# 代码中的类名、方法名、参数名已经指定，请勿修改，直接返回方法规定的值即可
#
# 
# @param heights int整型一维数组 
# @return int整型
#
class Solution:
    def largestRectangleArea(self , heights: List[int]) -> int:
        # write code here
        if not heights:
            return 0
        
        # 追加右侧零高哨兵以触发清栈
        extended_heights = heights + [0]
        # 初始化置入 -1 哨兵，省去后续判断栈空的操作
        monotonic_stack = [-1]
        max_area = 0
        
        for idx, current_h in enumerate(extended_heights):
            # 当遇到更矮的柱子，开始回溯结算
            while len(monotonic_stack) > 1 and current_h < extended_heights[monotonic_stack[-1]]:
                target_h = extended_heights[monotonic_stack.pop()]
                # 有了 -1 哨兵，宽度计算统一为当前索引减去新栈顶再减 1
                current_w = idx - monotonic_stack[-1] - 1
                
                computed_area = target_h * current_w
                if computed_area > max_area:
                    max_area = computed_area
                    
            monotonic_stack.append(idx)
            
        return max_area
```

## J 消防局的设立

```python
## add your code here
import sys

def optimize_fire_stations():
    tokens = sys.stdin.buffer.read().split()
    if not tokens:
        return
    
    total_nodes = int(tokens[0])
    if total_nodes == 1:
        print(1)
        return
    
    # 快速建立父节点追溯表
    parent_map = [0] * (total_nodes + 1)
    for node in range(2, total_nodes + 1):
        parent_map[node] = int(tokens[node - 1])
        
    INF = 10**9
    NEG_INF = -10**9
    
    near_station = [INF] * (total_nodes + 1)
    far_uncovered = [0] * (total_nodes + 1)
    station_count = 0
    
    # 自底向上反向处理：子节点将状态逆向推给父节点
    for node in range(total_nodes, 0, -1):
        # 阶段一：在结算前验证当前节点是否可被自身子树内部覆盖
        if near_station[node] + far_uncovered[node] <= 2:
            far_uncovered[node] = NEG_INF
            
        # 阶段二：距离已达阈值2，必须在当前节点强行建立消防局
        if far_uncovered[node] == 2:
            station_count += 1
            near_station[node] = 0
            far_uncovered[node] = NEG_INF
            
        # 阶段三：将当前节点精炼后的状态，向上推给其唯一的父节点
        if node > 1:
            p_node = parent_map[node]
            if near_station[node] + 1 < near_station[p_node]:
                near_station[p_node] = near_station[node] + 1
                
            if far_uncovered[node] + 1 > far_uncovered[p_node]:
                far_uncovered[p_node] = far_uncovered[node] + 1
                
    # 根节点残余未覆盖点的最终补网
    if far_uncovered[1] >= 0:
        station_count += 1
        
    print(station_count)

if __name__ == "__main__":
    optimize_fire_stations()
```
